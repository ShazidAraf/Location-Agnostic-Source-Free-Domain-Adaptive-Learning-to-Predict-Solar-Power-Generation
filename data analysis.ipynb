```python
# 50, 100/101, 


f3 = ['data/ca-pv-2006','data/ny-pv-2006','data/fl-pv-2006']

for g in f3:
    print(g,g[0:7])
    
    os.makedirs(g[0:7],exist_ok=True)
    
    f = sorted(os.listdir(g))
    
    for i in range(len(f)):
        f2 = copy.deepcopy(f[i]).split('_')
    #     print(f2)

        if f2[0]=='Actual' and f2[4]=='UPV' and f2[5]=='50MW':
            
            dataset = pd.read_csv(os.path.join(g,f[i]))
            data_numpy = dataset.to_numpy()
            p = data_numpy[:,1]            
            p = np.resize(p, (-1,6))
            p = np.mean(p,axis=1,keepdims=True)
            np.save(os.path.join(g[0:7],f[i][0:-4]+'.npy'),p)
```

```python
f3 = ['data/CA.csv','data/NY.csv','data/FL.csv']

for g in f3:
    
    dataset = pd.read_csv(os.path.join(g))
    data_numpy = dataset.to_numpy()
    np.save(os.path.join(g[0:7]+'_weather.npy'),data_numpy)
#     print(data_numpy.shape)
```

```python
import random
states = ['CA','FL','NY']



for i in range(len(states)):
      
    weather_feature = np.load('data/{0}_weather.npy'.format(states[i]),allow_pickle=True)
    power = np.load('data/{0}_power.npy'.format(states[i]),allow_pickle=True)
    power = np.resize(power,(-1,1))
    power = np.vstack((power,np.array([0])))
    data = np.hstack((weather_feature,power))
    
    test_idx = random.sample(range(weather_feature.shape[0]),int(0.2*weather_feature.shape[0]))
    train_idx = np.array(list(set(range(weather_feature.shape[0]))-set(test_idx)))
    
    train_data = data[train_idx]
    test_data = data[test_idx]
    
    np.save(os.path.join('data',states[i],'train.npy'),train_data)
    np.save(os.path.join('data',states[i],'test.npy'),test_data)
    
#     train
#     print(weather_feature.shape,power.shape)
```
